```python
import matplotlib.pyplot as plt
import numpy as np
```

#Data

```python
gender_labels = ['Male', 'Female', 'Non-binary', 'other']
gender_counts = [300, 400, 50, 20]

ages=np.random.randint(18, 70, size=1000)
```

#Bar chart for gender distribution

```python
plt.figure(figsize=(8, 6))
plt.bar(gender_labels, gender_counts, color='skyblue')
plt.title('Gender Distribution')
plt.xlabel('Gender')
plt.ylabel('Count')
plt.show()
```

#Histogram for age distribution

```python
plt.figure(figsize=(8, 6))
plt.hist(ages, bins=10, color='lightgreen', edgecolor='black')
plt.title('Age Distribution')
plt.xlabel('Age')
plt.ylabel('Frequency')
plt.show()
```
